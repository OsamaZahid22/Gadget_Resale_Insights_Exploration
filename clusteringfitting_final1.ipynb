```python
def calculate_kurtosis_skewness_for_columns(dataframe, columns):
    """
    Calculate kurtosis and skewness for specified numeric columns in a pandas DataFrame.

    Args:
    dataframe (pd.DataFrame): The dataframe to calculate kurtosis and skewness for.
    columns (list): A list of column names for which to calculate kurtosis and skewness.

    Returns:
    pd.DataFrame: A DataFrame with kurtosis and skewness for the specified columns.
    """
    kurtosis = dataframe[columns].kurtosis()
    skewness = dataframe[columns].skew()

    result_df = pd.DataFrame({'Kurtosis': kurtosis, 'Skewness': skewness})
    return result_df

# To use this function, first identify the numeric columns in your DataFrame.
numeric_columns = df.select_dtypes(include='number').columns.tolist()

# Then call the function with these columns
kurtosis_skewness_for_numeric_columns = calculate_kurtosis_skewness_for_columns(df, numeric_columns)

```

```python

def plot_device_brand_distribution(data):
    """
    Plot the distribution of device brands and their normalized used prices.

    Parameters:
    data (DataFrame): The pandas DataFrame containing the device data.

    Returns:
    None
    """
    
    fig, ax = plt.subplots(2, 1, figsize=(14, 12))

    sns.countplot(data=data, y='device_brand', order=data['device_brand'].value_counts().index, ax=ax[0])
    ax[0].set_title('Distribution of Device Brands')
    ax[0].set_xlabel('Count')
    ax[0].set_ylabel('Device Brand')

    brand_price = data.groupby('device_brand')['normalized_used_price'].mean().sort_values()
    brand_price.plot(kind='barh', ax=ax[1])
    ax[1].set_title('Average Normalized Used Price by Device Brand')
    ax[1].set_xlabel('Average Normalized Used Price')
    ax[1].set_ylabel('Device Brand')
    plt.tight_layout()
    plt.show()

def plot_correlation_matrix(data):
    """
    Plot the correlation matrix for numerical features in the dataset.

    Parameters:
    data (DataFrame): The pandas DataFrame containing the device data.

    Returns:
    None
    """
    numeric_data = data.select_dtypes(include=['float64', 'int64'])
    sns.set(style="white")
    corr = numeric_data.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    f, ax = plt.subplots(figsize=(11, 9))
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
    ax.set_title('Correlation Matrix of Numerical Features')
    plt.show()

```

```python
import matplotlib.pyplot as plt

def relation_graph(dataframe, column_x, column_y, title="Relation Graph", xlabel=None, ylabel=None, figsize=(10, 6)):
    """
    Create a scatter plot to visualize the relationship between two columns in a pandas DataFrame.

    Args:
    dataframe (pd.DataFrame): The dataframe containing the data.
    column_x (str): The name of the column to be plotted on the x-axis.
    column_y (str): The name of the column to be plotted on the y-axis.
    title (str, optional): The title of the graph.
    xlabel (str, optional): The label for the x-axis. If None, column_x name will be used.
    ylabel (str, optional): The label for the y-axis. If None, column_y name will be used.
    figsize (tuple, optional): The size of the figure.

    Returns:
    matplotlib.axes._subplots.AxesSubplot: The axes for the scatter plot.
    """
    plt.figure(figsize=figsize)
    plt.scatter(dataframe[column_x], dataframe[column_y])
    plt.title(title)
    plt.xlabel(xlabel if xlabel else column_x)
    plt.ylabel(ylabel if ylabel else column_y)
    plt.grid(True)
    return plt.gca()
```

```python
def labeled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    """

    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 2, 6))
    else:
        plt.figure(figsize=(n + 2, 6))

    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n],
    )

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage

    plt.show()  # show the plot
```

```python
def histogram_boxplot(data, feature, figsize=(15, 10), kde=False, bins=None):
    """
    Boxplot and histogram combined

    data: dataframe
    feature: dataframe column
    figsize: size of figure (default (15,10))
    kde: whether to show the density curve (default False)
    bins: number of bins for histogram (default None)
    """
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    sns.boxplot(
        data=data, x=feature, ax=ax_box2, showmeans=True, color="violet"
    )  # boxplot will be created and a triangle will indicate the mean value of the column
    sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins
    ) if bins else sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2
    )  # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color="green", linestyle="--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color="black", linestyle="-"
    )  # Add median to the histogram
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

from sklearn.metrics import silhouette_score

def elbow_method(data, range_of_clusters):
    """
    Perform the elbow method to determine the optimal number of clusters.

    Args:
    data (array-like): Scaled data for clustering.
    range_of_clusters (range): Range of cluster sizes to try.

    Returns:
    Inertia values and a plot of inertia across the range of cluster sizes.
    """
    inertias = []
    for k in range_of_clusters:
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(data)
        inertias.append(kmeans.inertia_)

    # Plot the inertia to find the elbow
    plt.figure(figsize=(8, 4))
    plt.plot(range_of_clusters, inertias, 'bo-')
    plt.xlabel('Number of clusters, k')
    plt.ylabel('Inertia')
    plt.title('Elbow Method For Optimal k')
    plt.grid(True)
    plt.show()
    return inertias

def silhouette_method(data, range_of_clusters):
    """
    Calculate and plot the Silhouette Score for determining the optimal number of clusters.

    Args:
    data (array-like): Scaled data for clustering.
    range_of_clusters (range): Range of cluster sizes to try.

    Returns:
    Silhouette scores and a plot of scores across the range of cluster sizes.
    """
    silhouette_scores = []
    for k in range_of_clusters:
        if k > 1:  # Silhouette score cannot be computed with fewer than 2 clusters
            kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
            kmeans.fit(data)
            score = silhouette_score(data, kmeans.labels_)
            silhouette_scores.append(score)

    # Plot the silhouette scores
    plt.figure(figsize=(8, 4))
    plt.plot(range_of_clusters[1:], silhouette_scores, 'ro-')  # Skip k=1 since no score is calculated
    plt.xlabel('Number of clusters, k')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Scores For Each k')
    plt.grid(True)
    plt.show()
    return silhouette_scores


```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def perform_kmeans(data_scaled, num_clusters=3):
    """
    Perform k-means clustering on scaled data.
    
    Parameters:
        data_scaled (numpy.ndarray): Scaled feature data for clustering.
        num_clusters (int): The number of clusters to form as well as the number of centroids to generate.
        
    Returns:
        KMeans: Trained KMeans instance.
        numpy.ndarray: Array of cluster labels.
    """
    kmeans = KMeans(n_clusters=num_clusters, n_init=10, random_state=42)
    cluster_labels = kmeans.fit_predict(data_scaled)
    return kmeans, cluster_labels

def plot_clusters(data_scaled, cluster_labels, kmeans, feature_names):
    """
    Plot the clusters along with the centroids on a 2D graph.
    
    Parameters:
        data_scaled (numpy.ndarray): Scaled feature data for clustering.
        cluster_labels (numpy.ndarray): Labels of each point.
        kmeans (KMeans): Trained KMeans instance.
        feature_names (list of str): Names of the features for axis labeling.
    """
    plt.figure(figsize=(10, 6))
    for cluster in set(cluster_labels):
        plt.scatter(data_scaled[cluster_labels == cluster, 0], data_scaled[cluster_labels == cluster, 1], 
                    s=50, label=f'Cluster {cluster+1}')
    
    # Plotting the centroids
    centroids = kmeans.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='yellow', label='Centroids')
    
    plt.title('Clusters of Devices')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.legend()
    plt.grid(True)
    plt.show()

def calculate_cluster_distribution(df, cluster_labels):
    """
    Calculate the distribution of data points in each cluster.
    
    Parameters:
        df (pandas.DataFrame): The dataframe with the original data.
        cluster_labels (numpy.ndarray): Labels of each point.
    
    Returns:
        pandas.DataFrame: DataFrame with the count and percentage of each cluster.
    """
    df['cluster'] = cluster_labels
    cluster_distribution = df['cluster'].value_counts().reset_index()
    cluster_distribution.columns = ['cluster', 'count']
    cluster_distribution['percentage'] = (cluster_distribution['count'] / len(df)) * 100
    return cluster_distribution


feature_names = ['normalized_used_price', 'normalized_new_price']
kmeans_model, labels = perform_kmeans(data_scaled)
plot_clusters(data_scaled, labels, kmeans_model, feature_names)
cluster_dist = calculate_cluster_distribution(df, labels)
print(cluster_dist)
```
